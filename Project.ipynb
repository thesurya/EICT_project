```python
img = cv2.imread('IMG_4949 (1).jpeg')
cv2_imshow(img)
#IMG_4949 (1)
#imgonlin
```

```python
#Detect faces using RetinaFace
img_faces = RetinaFace.detect_faces(img)
```

```python
#Iterate all objects key areas
for i in img_faces.keys():
  choose = img_faces[i]
  print(choose)
```

```python
for i in img_faces.keys():
  facial_parts = img_faces[i]
  #print(choose)
  recognize_face_area = facial_parts["facial_area"]
  cv2.rectangle(img, (recognize_face_area[2], recognize_face_area[3]), (recognize_face_area[0], recognize_face_area[1]), (255, 255, 255), 1)
```

```python
plt.figure(figsize= (10, 10))
plt.imshow(img[:, :, ::-1])
plt.show()
```

```python
#Extract individual faces from the image
import matplotlib.pyplot as plt
ext_faces = RetinaFace.extract_faces(img_path = "IMG_4949 (1).jpeg", align = True)
```

```python
#from PIL import Image 
i = 1
for face in ext_faces:
  plt.imshow(face)
  #pil_img = Image.fromarray(face)
  #pil_img.save("img"+str(1)+".jpg")
  plt.savefig("pic_"+str(i)+".jpeg")
  plt.show()
  #files.download("pic_"+str(i)+".png")
  i+=1 
```
